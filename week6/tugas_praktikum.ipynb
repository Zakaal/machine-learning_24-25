```python
best_dt = grid_search_dt.best_estimator_
best_dt.fit(X_train, y_train)

y_pred_dt = best_dt.predict(X_test)
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print("Test set accuracy for Decision Tree: {:.2f}".format(accuracy_dt))
```

**Model Training & Model prediction (AdaBoost)**

```python
best_ab = grid_search_ab.best_estimator_
best_ab.fit(X_train, y_train)

y_pred_ab = best_ab.predict(X_test)
accuracy_ab = accuracy_score(y_test, y_pred_ab)
print("Test set accuracy for AdaBoost: {:.2f}".format(accuracy_ab))
```

**Hasil**

# **TUGAS 3**

**Import library**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

**Load Dataset**

```python
from google.colab import drive
drive.mount('/content/drive')

dbt = pd.read_csv('/content/drive/MyDrive/machine-leaning/pertemuan-5/dataset/diabetes1.csv')

dbt.head()
```

```python
feature_columns = dbt.columns[:-1]  # semua kolom kecuali 'Outcome'
imputer = SimpleImputer(missing_values=0, strategy='mean')
dbt[feature_columns] = imputer.fit_transform(dbt[feature_columns])
```

```python
# Fitur dan label
X = dbt[feature_columns]
y = dbt['Outcome']
```

**Split Data**

```python
# Memisahkan data menjadi training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)
```

**Hyperparameter Tuning**

```python
# Tuning hyperparameter untuk Logistic Regression
logreg = LogisticRegression(solver='liblinear')

param_grid_logreg = {
    'C': [0.01, 0.1, 1, 10, 100]
}
grid_logreg = GridSearchCV(logreg, param_grid_logreg, cv=5)
grid_logreg.fit(X_train_std, y_train)

best_logreg = grid_logreg.best_estimator_

# Tuning hyperparameter untuk SVM
svm_poly = SVC(kernel='poly')

param_grid_svm = {
    'C': [0.01, 0.1, 1, 10, 100],
    'degree': [2, 3, 4]
}
grid_svm = GridSearchCV(svm_poly, param_grid_svm, cv=5)
grid_svm.fit(X_train_std, y_train)

best_svm = grid_svm.best_estimator_

# Tuning hyperparameter untuk Decision Tree
dtree = DecisionTreeClassifier()

param_grid_dtree = {
    'max_depth': [None, 5, 10, 20],
    'min_samples_split': [2, 5, 10]
}
grid_dtree = GridSearchCV(dtree, param_grid_dtree, cv=5)
grid_dtree.fit(X_train, y_train)

best_dtree = grid_dtree.best_estimator_
```

**Voting Classifier**

```python
# Membuat Voting Classifier
voting = VotingClassifier(
    estimators=[
        ('logreg', best_logreg),
        ('svm_poly', best_svm),
        ('decision_tree', best_dtree)
    ],
    voting='hard'
)

# Melatih model voting
voting.fit(X_train_std, y_train)
```

**Evaluasi Model**

```python
y_pred_voting = voting.predict(X_test_std)
accuracy = accuracy_score(y_test, y_pred_voting)
print("Akurasi Voting Classifier: {:.2f}".format(accuracy))
```
